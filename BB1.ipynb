```python
# Bully Block App

import tkinter as tk
import tkinter.ttk as ttk
from tkinter import Toplevel

# @Shivank - you will need to comment out this line and replace it with "from tk import Button" for running on Windows.
from tkmacosx import Button
from tkinter import messagebox as m_box


import random

#Creating the windows
win = tk.Tk()
win.title("Bully Block")
win.geometry("450x750")

win.configure(bg='black')
win.title("Bully Block")
win.geometry("500x750")
#Set buttons to fix for window size
win.resizable(False, False)

#Create an empty array of strings
friends = []
parents = []
teachers = []


win_child = None
name_text = None
phone_text = None
type_combo = None

#Create a label saying BullyBlock at 175, 25 with a font size of 40
label1 = tk.Label(win, text="BullyBlock", font=("Arial Bold", 40), fg='gray', bg='black')
label1.place(x=150, y=10)


button1 = Button(win, text="Verbal abuse/Mocking", width=150, height=80,  bg='yellow', fg='black', activebackground='brown', activeforeground='black', )
button1.place(x=175, y=75)
button2=Button(win, text="Pushing/Shoving", width=150, height=80, bg='orange', fg='black', activebackground='brown', activeforeground='black', )
button2.place(x=175, y=150)
button3=Button(win, text="Fighting", width=150, height=80, bg='red', fg='black', activebackground='brown', activeforeground='black', )
button3.place(x=175, y=225)
button4=Button(win, text="Guns/kidnapping", width=150, height=80, bg='black', fg='white', activebackground='brown', activeforeground='black', )
button4.place(x=175, y=300)

#Create a button at 25,75 saying settings and when clicked, it will open a new window
button5 = Button(win, text="Settings", width=75, height=40, bg='gray', fg='black', activebackground='brown', activeforeground='black')
button5.place(x=50, y=375)
button6 = Button(win, text="Add", width=75, height=40, bg='gray', fg='black', activebackground='brown', activeforeground='black')
button6.place(x=355, y=375)
button7 = Button(win, text="Sign In", width=75, height=40, bg='gray', fg='black', activebackground='brown', activeforeground='black')
button7.place(x=205, y=475)
button8 = Button(win, text="Sign Up", width=75, height=40, bg='gray', fg='black', activebackground='brown', activeforeground='black')
button8.place(x=205, y=550)

#Create a function called random that will get the text from text 1 and will show on button10 when clicked
def friends2_func():
    m_box.showinfo("Friends", friends)

#Defining the functions
def button1_func():
    m_box.showinfo("Level 1", "Verbal abuse/Mocking")

def button2_func():
    m_box.showinfo("Level 2", "Physical abuse such as shoving")

def button3_func():
    m_box.showinfo("Level 3", "Physical abuse such as getting into fights")

def button4_func():
    m_box.showinfo("Level 4", "Guns/kidnapping")

def button5_func():
    win_child=Toplevel(win) # Child window 
    win_child.geometry("450x750")  # Size of the window 
    win_child.title("Settings")
    #Create a label in child window saying Profile
    label2 = tk.Label(win_child, text="Settings", font=("Arial Bold", 20), fg='gray', bg='black')
    label2.place(x=175, y=15)

def button6_func():
    win_child=Toplevel(win) # Child window 
    win_child.geometry("450x750")  # Size of the window 
    win_child.title("Add")
    #Create a label in child window saying Profile
    label3 = tk.Label(win_child, text="Add friends", font=("Arial Bold", 20), fg='gray', bg='black')
    label3.place(x=175, y=15)
    #Make a text box saying " Enter the phone number of the person you want to add"
    makeScreen (win_child)


def button7_func():
    win_child=Toplevel(win) # Child window 
    win_child.geometry("450x750")  # Size of the window 
    win_child.title("Sign In")
    #Create a label in child window saying Profile
    label4 = tk.Label(win_child, text="Sign In", font=("Arial Bold", 20), fg='gray', bg='black')
    label4.place(x=175, y=15)

def button8_func():
    win_child=Toplevel(win) # Child window 
    win_child.geometry("450x750")  # Size of the window 
    win_child.title("Sign Up")
    #Create a label in child window saying Profile
    label5 = tk.Label(win_child, text="Sign Up", font=("Arial Bold", 20), fg='gray', bg='black')
    label5.place(x=175, y=15)

def makeScreen(childWin):
    #Create a label in child window saying Profile
    label6 = tk.Label(childWin, text="Enter the phone number of the person you want to add", font=("Arial Bold", 10), fg='gray', bg='black')
    label6.place(x=50, y=50)
    #Create a text box in child window
    name_text = tk.Text(childWin, height=2, width=20)
    name_text.place(x=50, y=100)
    phone_text = tk.Text(childWin, height=2, width=20)
    phone_text.place(x=50, y=150)
    type_combo = ttk.Combobox(master=childWin, width=60, textvariable=tk.StringVar())
    type_combo.config(state='readonly')
    type_combo['values'] = ('Friend', 'Parent', 'Teacher')
    type_combo.current(0)
    type_combo.place(x=50, y=200)
    #Create a button in child window
    addbutton = Button(childWin, text="Add", width=75, height=40, bg='gray', fg='black', activebackground='brown', activeforeground='black')
    addbutton.place(x=50, y=400)
    addbutton.config(command = add_friends_func)
    #Make a button that shows your friends whe clicked
    showButton = Button(childWin, text="Show Friends", width=200, height=40, bg='gray', fg='black', activebackground='brown', activeforeground='black')
    showButton.place(x=50, y=450)
    showButton.config(command = show_friends_func)

    
def add_friends_func():
    name = name_text.text
    phone = phone_text.text
    type = type_combo.get()
    if (name.length > 0 and phone.length > 0 and type.length > 0):
        if (type == 'Friend'):
            friends.append(name)
        elif (type == 'Parent'):
            parents.append(name)
        else:
            teachers.append(name)

def show_friends_func():
    m_box.showinfo("Friends", friends)

#Calling the functions
button1.config(command = button1_func)
button2.config(command = button2_func)
button3.config(command = button3_func)
button4.config(command = button4_func)
button5.config(command = button5_func)
button6.config(command = button6_func)
button7.config(command = button7_func)
button8.config(command = button8_func)


#Running the mainloop
win.mainloop()
























```
