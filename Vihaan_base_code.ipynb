```python
# create an app with 4 buttons
# 1. button1 says "level 1"
# 2. button2 says "level 2"
# 3. button3 says "level 3"
# 4. button4 says "level 4"
# button one is white
# button two is yellow
# button three is orange
# button four is red
# when you click button1, the screen shows a yellow background with the text "mocking"
# when you click button2, the screen shows an orange background with the text "shoving"
# when you click button3, the screen shows a red background with the text "fighting"
# when you click button4, the screen shows a black background with the text "weapons"
# each button is in the spae of a rounded rectangle
# each button has a White border
# each button has a black text
# each button has a black shadow
# the backround of the app is light blue
# the text of the app is black
# the app is called "Bully Block"
# use tkinter





from tkinter import *
from tkinter import messagebox
import tkinter as tk
import webbrowser
# create a tkinter window
root = tk.Tk()
# create a title for the window
root.title("Bully Block")
# create a label for the window
label = tk.Label(root, text="BullyBlock", font=("Arial Bold", 50))
label.grid(column=0, row=0)
# change postion of the label
label.place(x=70, y=40)
# change color of the label
label.config(bg="light blue")
# change color of the text
label.config(fg="black")


# Add another label 
label1 = tk.Label(root, text="Hello <Users First Name>", font=("Arial Bold", 20))
label1.grid(column=0, row=1)
# change position of the label
label1.place(x=80, y=150)
# change higlight color of the label
label1.config(bg="light blue")
# create a function for the buttons
def clicked1():
    messagebox.showinfo("Level 1", "message friends and family")
def clicked2():
    messagebox.showinfo("Level 2", "message friends, family, and teachers")
def clicked3():
    messagebox.showinfo("Level 3", "message friends, family, teachers, and surrounding people")
def clicked4():
    messagebox.showinfo("Level 4", "message friends, family, teachers, surrounding people, and police")
# create a button for the window
button1 = tk.Button(root, text="Level 1", bg="white", fg="black", command=clicked1)
button1.grid(column=0, row=1)
button2 = tk.Button(root, text="Level 2", bg="yellow", fg="black", command=clicked2)
button2.grid(column=0, row=2)
button3 = tk.Button(root, text="Level 3", bg="orange", fg="black", command=clicked3)
button3.grid(column=0, row=3)
button4 = tk.Button(root, text="Level 4", bg="red", fg="black", command=clicked4)
button4.grid(column=0, row=4)
# change size of the buttons
button1.config(height=5, width=20)
button2.config(height=5, width=20)
button3.config(height=5, width=20)
button4.config(height=5, width=20)
# change shape of the buttons to rounded rectangle
button1.config(relief="ridge", borderwidth=5)
button2.config(relief="ridge", borderwidth=5)
button3.config(relief="ridge", borderwidth=5)
button4.config(relief="ridge", borderwidth=5)
# change positon of the buttons
button1.place(x=90, y=200)
button2.place(x=90, y=300)
button3.place(x=90, y=400)
button4.place(x=90, y=500)
# change the color of buttons
button1.config(bg="white")
button2.config(bg="yellow")
button3.config(bg="orange")
button4.config(bg="red")


#create a button to open a new window to add contacts
def clicked5():
    window = tk.Toplevel(root)
    window.title("Add Contact")
    window.geometry("400x700")
    window.configure(bg="light blue")
    label = tk.Label(window, text="Add Contact", font=("Arial Bold", 50))
    label.grid(column=0, row=0)
    label.place(x=50, y=50)
    label.config(bg="light blue")
    label.config(fg="black")
    name = tk.Label(window, text="Name", font=("Arial Bold", 20))
    name.grid(column=0, row=1)
    name.place(x=165, y=150)
    name.config(bg="light blue")
    name.config(fg="black")
    phonenumber = tk.Label(window, text="Phone Number", font=("Arial Bold", 20))
    phonenumber.grid(column=0, row=2)
    phonenumber.place(x=125, y=250)
    phonenumber.config(bg="light blue")
    phonenumber.config(fg="black")
    email = tk.Label(window, text="Email", font=("Arial Bold", 20))
    email.grid(column=0, row=3)
    email.place(x=165, y=350)
    email.config(bg="light blue")
    email.config(fg="black")
    entry = tk.Entry(window, width=20)
    entry.grid(column=1, row=1)
    entry.place(x=100, y=200)
    entry.config(bg="white")
    entry.config(fg="black")
    entry1 = tk.Entry(window, width=20)
    entry1.grid(column=1, row=2)
    entry1.place(x=100, y=300)
    entry1.config(bg="white")
    entry1.config(fg="black")
    entry2 = tk.Entry(window, width=20)
    entry2.grid(column=1, row=3)
    entry2.place(x=100, y=400)
    entry2.config(bg="white")
    entry2.config(fg="black")
    # create a drop down menu with the options "friends", "family", and "teachers"
    clicked = tk.StringVar()
    clicked.set("Contact Type")
    drop = tk.OptionMenu(window, clicked, "Friend", "Family", "Teacher")
    drop.grid(column=0, row=4)
    drop.place(x=135, y=470)
    drop.config(bg="light blue")
    drop.config(fg="Black")
    # change font size of the drop down menu
    drop.config(font=("Arial Bold", 20))
    # change the color of the drop down menu
    drop.config(bg="white")
    drop.config(fg="black")
    # function to save contact information to a file
    def save_contact(name, phone, email):
        with open('contacts.txt', 'a') as file:
            file.write(f"{name}, {phone}, {email}\n")
    # create a button to save the contact information
    button = tk.Button(window, text="Save", bg="white", fg="black", command=lambda: save_contact(entry.get(), entry1.get(), entry2.get()))
    button.grid(column=0, row=5)
    button.place(x=150, y=550)
    button.config(height=1, width=8)
    button.config(relief="ridge", borderwidth=0.1)
    button.config(bg="white")
    # create a button to delete the contact information
    def delete_contact():
        with open('contacts.txt', 'w') as file:
            file.write("")
    delete_contact = tk.Button(window, text="Delete", bg="white", fg="black", command=delete_contact)
    delete_contact.grid(column=1, row=5)
    delete_contact.place(x=150, y=575)
    delete_contact.config(height=1, width=8)
    delete_contact.config(relief="ridge", borderwidth=0.1)
    delete_contact.config(bg="white")
    # create a button to see the contact information
    def see_contacts():
        with open('contacts.txt', 'r') as file:
            messagebox.showinfo("Contacts", file.read())
    button1 = tk.Button(window, text="See Contacts", bg="white", fg="black", command=see_contacts)
    button1.grid(column=1, row=5)
    button1.place(x=150, y=600)
    button1.config(height=1, width=8)
    button1.config(relief="ridge", borderwidth=0.1)
    button1.config(bg="white")

    entry


# create a button for the window
button5 = tk.Button(root, text="Add Contacts", bg="white", fg="black", command=clicked5)
# change the size of the button
button5.config(height=1, width=8)
# change the shape of the button
button5.config(relief="ridge", borderwidth=0.1)
# change the position of the button
button5.place(x=10, y=10)
# change the color of the button
button5.config(bg="white")
# create a function for the link
def callback(url):
    webbrowser.open_new(url)
# create a link for the window
link = Label(root, text="Click here to learn more about bullying", fg="black", cursor="hand2")
link.grid(column=0, row=5)
link.bind("<Button-1>", lambda e: callback("https://www.stopbullying.gov/what-is-bullying"))
# change the position of the link
link.place(x=80, y=625)
# change backround color of the link
link.config(bg="light blue")
# create a function for the window
def clicked():
    messagebox.showinfo("Bully Block", "Thank you for using Bully Block")
# change the background color of the window to light blue
root.configure(bg="light blue")
# change the size of the window
root.geometry("400x700")
# stop creating more windows
root.mainloop()

# end of code
```
